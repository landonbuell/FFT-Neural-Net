```python
def Read_CSV(filename,skip=1):
    """ Read Data From CSV File into pandas dataframe """
    data = np.loadtxt(filename,dtype=float,delimiter='\t',
                      skiprows=skip)
    return data

intdir = 'C:/Users/Landon/Documents/GitHub/FFT-Neural-Net/Binary_Classifier_v1'
os.chdir(intdir)

target = Read_CSV('is_string.txt',skip=0)

print("Loaded Target Data:",len(target),"points")
```

```python
L_FFT_Frame = pd.read_csv('L_FFT_Frame.txt',sep='\t',index_col=0)
L_FFT_Frame = L_FFT_Frame.transpose()

L_FFT_Frame.head()
```

```python
        #### Break into Train/Test Data set ####
state = 14
L_FFT = np.array(L_FFT_Frame)
target = np.array(target,dtype=bool)

X_train,X_test = model.train_test_split(L_FFT_Frame,test_size=0.1,random_state=state)
Y_train,Y_test = model.train_test_split(target,test_size=0.1,random_state=state)

X_test_arr = np.array(X_test)

print("Length of Training Data:",len(X_train))
print("Length of Testing Data:",len(X_test))
```

```python
        #### Train the Classifier ####
    
string_clf = SGDClassifier(random_state=state)
string_clf.fit(X_train,Y_train)

X_test.head()
```

```python
            #### Test Classifier ####

idx = [10,20,32,43,51,83,98]

print("Is String Instrument:")
print(Y_test[idx])

print('Classifier Prediction:')
for i in idx:
    print(string_clf.predict([X_test_arr[i]]))
```

```python
            #### Create Confusion Matrix ####

Y_pred = model.cross_val_predict(string_clf,X_train,Y_train,cv=2)
conf_mat = metrics.confusion_matrix(Y_train,Y_pred)

print(conf_mat)
print("\n")
```

```python
            #### Othert Metrics ####

print("True Negatives:",conf_mat[0][0])
print("False Negatives:",conf_mat[1][0])
print("False Positives:",conf_mat[0][1])
print("True Positives:",conf_mat[1][1])

""" This is NOT a very good classifier - Need to improve Volume of Data Set!!!!"""
print("\n")
print("Number of Strings in training Set:",len(np.where(Y_train == True)[0]))
print("Non Strings in training Set:",len(np.where(Y_train == False)[0]))
```

```python
print("Precision Score:")
prec = metrics.precision_score(Y_train,Y_pred)
print("Correctly Find String Intrument",100*prec,"% of time")

print("Recall Score:")
recl = metrics.recall_score(Y_train,Y_pred)
print("Find String",100*recl,"% of time")
```
