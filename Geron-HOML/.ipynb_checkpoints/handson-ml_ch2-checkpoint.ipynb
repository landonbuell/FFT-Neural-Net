```python
def fetch_housing_data(housing_url=HOUSING_URL, housing_path=HOUSING_PATH):
    """ From Geron, pg. 44 - Collects Housing data from online """
    os.makedirs(housing_path, exist_ok=True)
    tgz_path = os.path.join(housing_path, "housing.tgz")
    urllib.request.urlretrieve(housing_url, tgz_path)
    housing_tgz = tarfile.open(tgz_path)
    housing_tgz.extractall(path=housing_path)
    housing_tgz.close()

def load_housing_data (path=HOUSING_PATH):
    """ From Geron, pg. 44 - Reads Housing data from downloaded file """
    csv_path = os.path.join(path,'housing.csv')
    return pd.read_csv(csv_path)

# collect data & read into dataframe
fetch_housing_data()
housing = load_housing_data()
```

```python
""" Create a Data Testing Set """

# to make this notebook's output identical at every run
np.random.seed(42)

def split_train_test(data,test_ratio):
    """ From Geron, pg. 49 - split data into training/testing """
    # shuffle data rows around
    shuffled_indicies = np.random.permutation(len(data))
    # ratio of train/test data
    test_set_size = int(len(data)*test_ratio)
    # slice array accordingly
    test_indicies = shuffled_indicies[:test_set_size]
    train_indicies = shuffled_indicies[test_set_size:]
    # return separate DataFrames [iloc is integer based indexing]
    return data.iloc[train_indicies],data.iloc[test_indicies]

# split data w/ function above
train_set,test_set = split_train_test(housing,0.2)

# print legnth of sets
print("Training Set Length:",len(train_set))
print("Testing Set Length:",len(test_set))
```

```python
# add index column to housing DF
#housing_with_id = housing.reset_index()

# use sklearn to split dataset
train_set,test_set = sk.model_selection.train_test_split(housing,test_size=0.2,random_state=42)

# create new  column in DF
housing["income_cat"] = pd.cut(housing["median_income"],
                              bins=[0.0,1.5,3.0,4.5,6.0,np.inf],
                              labels=[1,2,3,4,5])
# Now housing is also group by categorizing incomes into 5 bins
housing["income_cat"].hist()
plt.show()
```

```python
# stratified sampling based on income category
split = sk.model_selection.StratifiedShuffleSplit(n_splits=1,
                                         test_size=0.2,random_state=42)
# split is a pandas class object
for train_index,test_index in split.split(housing,housing['income_cat']):
    strat_train_set = housing.loc[train_index]
    strat_test_set = housing.loc[test_index]
    
# check income proportions in test set:
print(strat_test_set["income_cat"].value_counts()/len(strat_test_set))
```

```python
# remove the 'income_cat' column
for set_ in (strat_train_set,strat_test_set):
    set_.drop("income_cat",axis=1,inplace=True)
```

```python
""" Visualize Data for Insight """

housing = strat_train_set.copy()
# produce scatter plot to visualize geographic data
housing.plot(kind='scatter',x='longitude',y='latitude',
             alpha=0.4,s=housing['population']/100,label='population',
            figsize=(12,8),c='median_house_value',
             cmap=plt.get_cmap('jet'),colorbar=True)
plt.legend()
plt.show()

#now we can easily see groups of data
    # i.e. price by color, population by size
    # general geography as well
```
