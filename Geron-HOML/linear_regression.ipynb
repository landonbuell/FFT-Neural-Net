```python
# Create an array of x-value data
x = np.arange(-2,+2,0.001)

# create relationships between data

def linear_func (m,b,x):
    """ returns linear relationship between x & y """
    return m*x + b

def quadratic_func(a,b,c,x):
    """ Returns quadratic relationship between x & y """
    return a*x**2 + b*x + c
```

```python
# create a relationships
m,d = 2,0.5
lin = linear_func(m,d,x)
a,b,c = -3,2,4
quad = quadratic_func(a,b,c,x)
```

```python
# Add some noise to data

def add_noise (arr):
    """ Add some 'noise' to an array of floats """
    vals = np.random.normal(-1,1,size=len(arr))
    return arr + vals

# add the noise
lin_noise = add_noise(lin)
quad_noise = add_noise(quad)
```

```python
# Apply Linear Regression

def linear_regression (xdata,ydata,order):
    """ Apply nth order linear regression to dataset """
    coeffs = np.polyfit(xdata,ydata,deg=order)
    fitdata = np.polyval(coeffs,xdata)
    return fitdata,coeffs

# Fit datasets
lin_fit,lin_coeffs = linear_regression(x,lin_noise,2)
quad_fit,quad_coeffs = linear_regression(x,quad_noise,3)
```

```python
plt.figure(figsize=(20,8))
plt.title("Regressed Relationships",size=40,weight='bold')
plt.xlabel("X - Axis",size=20,weight='bold')
plt.ylabel("Y - Axis",size=20,weight='bold')

plt.plot(x,lin,label='Linear Relationship')
plt.plot(x,lin_noise,label="Linear w/ Noise")
plt.plot(x,lin_fit,label="Linear w/ Fit")

plt.legend()
plt.grid()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(20,8))
plt.title("Regressed Relationships",size=40,weight='bold')
plt.xlabel("X - Axis",size=20,weight='bold')
plt.ylabel("Y - Axis",size=20,weight='bold')

plt.plot(x,quad,label='Quadratic Relationship')
plt.plot(x,quad_noise,label="Quadratic w/ Noise")
plt.plot(x,quad_fit,label="Quadratic w/ Fit")

plt.legend()
plt.grid()
plt.tight_layout()
plt.show()
```
